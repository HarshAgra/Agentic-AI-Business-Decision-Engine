```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()

client = AzureOpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    api_version="2024-02-15-preview",
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT")
)
```

```python
def call_llm(prompt):
    response = client.chat.completions.create(
        model=os.getenv("AZURE_OPENAI_DEPLOYMENT"),
        messages=[
            {"role": "system", "content": prompt}
        ],
        temperature=0.3
    )
    
    return response.choices[0].message.content
```

```python
def market_agent(user_input):
    prompt = f"""
    You are a Market Analyst AI.

    Analyze the following:

    {user_input}

    Output:
    Demand:
    Competition:
    Customer Fit:
    Opportunity Score:
    Reason:
    """
    
    return call_llm(prompt)
```

```python
result = market_agent(
    "Product: Budget Smartphone, Market: Tier-2 Cities, Segment: Price-sensitive"
)

print(result)
```

```python
def financial_agent(market_output, price, cost):
    prompt = f"""
    You are a Financial Analyst AI.

    Market Insights:
    {market_output}

    Selling Price: {price}
    Estimated Cost: {cost}

    Tasks:
    1. Estimate revenue level (Low/Medium/High)
    2. Estimate profit margin (%)
    3. Determine ROI level (Low/Medium/High)
    4. Financial viability (Good / Moderate / Poor)

    Output:
    Revenue:
    Profit Margin:
    ROI:
    Financial Viability:
    Reason:
    """
    
    return call_llm(prompt)
```

```python
financial_result = financial_agent(
    market_output=result,
    price=12000,
    cost=8000
)

print(financial_result)
```

```python
def risk_agent(market_output):
    prompt = f"""
    You are a Risk Analyst AI.

    Market Insights:
    {market_output}

    Tasks:
    1. Determine risk level (Low/Medium/High)
    2. Identify key risks
    3. Suggest mitigation

    Output:
    Risk Level:
    Key Risks:
    Mitigation:
    Reason:
    """
    
    return call_llm(prompt)
```

```python
risk_result = risk_agent(result)

print(risk_result)
```

```python
def decision_agent(market, financial, risk):
    prompt = f"""
    You are a Strategy Head AI.

    Your job is to make a final business decision.

    Inputs:

    Market Analysis:
    {market}

    Financial Analysis:
    {financial}

    Risk Analysis:
    {risk}

    Decision Rules:
    - If ROI is Low OR Financial Viability is Poor → Do Not Launch
    - If Risk Level is High → Launch with Caution
    - If Demand is High AND ROI is High → Strongly Recommend Launch

    Tasks:
    1. Analyze all inputs
    2. Apply decision rules
    3. Give final recommendation

    Output:
    Final Decision:
    Confidence Level:
    Reason:
    """
    
    return call_llm(prompt)
```

```python
def run_workflow(user_input, price, cost):

    market_output = market_agent(user_input)

    financial_output = financial_agent(market_output, price, cost)

    risk_output = risk_agent(market_output)

    decision_output = decision_agent(
        market_output,
        financial_output,
        risk_output
    )

    return {
        "Market": market_output,
        "Financial": financial_output,
        "Risk": risk_output,
        "Decision": decision_output
    }
```

```python
result = run_workflow(
    user_input="Product: Budget Smartphone, Market: Tier-2 Cities, Segment: Price-sensitive",
    price=12000,
    cost=8000
)

for key, value in result.items():
    print(f"\n--- {key} ---\n")
    print(value)
```
